```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
lista_articulos = pd.Series(articulos)
lista_compra = lista_articulos[compra]

print(lista_compra)

def precio_total(lista_compra):
    suma = lambda x, y: x + y
    total = reduce(suma, lista_compra)
    print("El precio total es: ", total)

# B. Una función que calcule el precio total de la compra con un descuento del 10%

def precio_total_descuento_10(lista_compra):
    total = reduce(lambda x,y: x+y, lista_compra) * 0.9
    print("El precio total con un 10% de descuento es: ", total)

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def descuento_si_mas_de_100(lista_compra):
    total = reduce(lambda x,y: x+y, lista_compra)
    if total > 100:
        total *= 0.9
        print("El precio total con un 10% de descuento es: ", total)
    else:
        print("El precio total es: ", total)


# D. Una función que calcule el precio total aplicando el IVA (21%)
def aplica_IVA(lista_compra):
    total_mas_IVA = round(reduce(lambda x,y: x+y, lista_compra) * 1.21, 2)
    print("El total de la compra con un IVA del 21% es: ", total_mas_IVA)


# E.lista los artículos cuyo precio es superior a 20€
articulos_mayores_20 = lista_articulos[lista_articulos > 20]
articulos_mayores_20
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]

#lo tranformo en un dataframe
lista_alumnos= pd.DataFrame(alumnos, columns=["Nombre", "Apellido", "Curso", "Notas"])


def prom_alumnos_curso(curso):
    
    #itero la lista de tuplas, y transformo las notas en sus promedios.
    lista_1 = [] 
    for indice, item in enumerate(alumnos):
        prom = reduce(lambda x,y: x+y, item[3])/len(item[3])
        lista_1.append(prom)
    lista_alumnos["Notas"] = lista_1
    
    #hago un dataframe nuevo por cada curso
    curso_1 = lista_alumnos[lista_alumnos["Curso"] == "1"]
    curso_2 = lista_alumnos[lista_alumnos["Curso"] == "2"]

    #le digo que hacer, segun el curso que se elige
    if curso == 1:
        print(curso_1[["Nombre", "Apellido", "Notas"]])
    elif curso == 2:
        print(curso_2[["Nombre", "Apellido", "Notas"]])

#pruebo la funcion
print(prom_alumnos_curso(2))
```
